```
import matplotlib.pyplot as plt
import numpy as np

# -------------------------------------------------------
# DATA FROM YOUR EXPERIMENTS (Copied from your logs)
# -------------------------------------------------------

# Experiment A: "The color of a square circle..." (CONFUSION)
# These are the entropy values from your first log
entropy_confused = [5.53, 6.33, 4.82, 5.29, 5.26, 3.59, 4.15, 4.20, 5.63, 6.42]

# Experiment B: "The sun rises in the..." (CLARITY)
# These are the entropy values from your second log
# Note how they drop drastically after the first few words
entropy_focused = [5.57, 3.44, 4.93, 2.33, 0.45, 0.02, 0.01, 0.01, 0.02, 0.01]

# -------------------------------------------------------
# VISUALIZATION CODE
# -------------------------------------------------------
plt.figure(figsize=(10, 6))

# Plot the "Confused" line in Red
plt.plot(entropy_confused, marker='o', linestyle='-', color='red', linewidth=2, label='Confused Model (Hallucinating)')

# Plot the "Focused" line in Blue
plt.plot(entropy_focused, marker='o', linestyle='--', color='blue', linewidth=2, label='Focused Model (Fact-Based)')

# Add a "Threshold" line (The Sakshi Limit)
plt.axhline(y=3.5, color='gray', linestyle=':', label='Sakshi Intervention Threshold (3.5)')

# Labels and Title
plt.title('The Sakshi Effect: Entropy Stabilization Analysis', fontsize=14)
plt.xlabel('Token Generation Step', fontsize=12)
plt.ylabel('Cognitive Entropy (Uncertainty)', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)

# Show the graph
plt.show()
```
